```python
class telo():
    def __init__(self, m, x, y, z, vx, vy, vz, ax, ay, az, marker):
        self.m = m
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.ax = ax
        self.ay = ay
        self.az = az
        self.marker = marker


def jedinice_masa(m):
    return m * (2.325e9)

def jedinice_brzina(v):
    return v * 3.24e-15
  
def rastojanje(t1,t2):
    return math.sqrt((t1.x-t2.x)**2+(t1.y-t2.y)**2 + (t1.z-t2.z)**2)

def main():
    
    dt = 1e7 * 365 * 86400
    G = 4.53e-39
    tmax = 5e9 * 365 * 86400
    t = 0

    tela = []
    
    for i in range(200):
        tela.append(telo(jedinice_masa(m_h[i]), x_h[i], y_h[i], z_h[i],jedinice_brzina(vx_h[i]), jedinice_brzina(vy_h[i]), jedinice_brzina(vz_h[i]), 0, 0, 0, "halo"))
    for i in range(100):
        tela.append(telo(jedinice_masa(m_d[i]), x_d[i], y_d[i], z_d[i], jedinice_brzina(vx_d[i]), jedinice_brzina(vy_d[i]), jedinice_brzina(vz_d[i]), 0, 0, 0, "disk")) 
    for i in range(100):
        tela.append(telo(jedinice_masa(m_b[i]), x_b[i], y_b[i], z_b[i], jedinice_brzina(vx_b[i]), jedinice_brzina(vy_b[i]), jedinice_brzina(vz_b[i]), 0, 0, 0, "bulge"))

    for i in range(200):
        tela.append(telo(jedinice_masa(m_h2[i]), x_h2[i], y_h2[i], z_h2[i], jedinice_brzina(vx_h2[i]), jedinice_brzina(vy_h2[i]), jedinice_brzina(vz_h2[i]), 0, 0, 0, "halo"))
    for i in range(100):
        tela.append(telo(jedinice_masa(m_d2[i]), x_d2[i], y_d2[i], z_d2[i], jedinice_brzina(vx_d2[i]), jedinice_brzina(vy_d2[i]), jedinice_brzina(vz_d2[i]), 0, 0, 0, "disk")) 
    for i in range(100):
        tela.append(telo(jedinice_masa(m_b2[i]), x_b2[i], y_b2[i], z_b2[i], jedinice_brzina(vx_b2[i]), jedinice_brzina(vy_b2[i]), jedinice_brzina(vz_b2[i]), 0, 0, 0, "bulge"))


    x1_h = []
    y1_h = []
    z1_h = []

    x1_d = []
    y1_d = []
    z1_d = []
    
    x1_b = []
    y1_b = []
    z1_b = []

    N = len(tela)
    print(len(tela))
    
    while(t<tmax):
    
        for i in range(N):
            for j in range(N):

                if (i!=j):

                    d = rastojanje(tela[i],tela[j])

                    #if i == 0 and j == 1:
                    #    print(i, j, d)

                            
                    tela[i].ax += -G * tela[j].m * (tela[i].x-tela[j].x) / ((d)**3)
                    tela[i].ay += -G * tela[j].m * (tela[i].y-tela[j].y) / ((d)**3)
                    tela[i].az += -G * tela[j].m * (tela[i].z-tela[j].z) / ((d)**3)


        for i in range(N):
            tela[i].vx += dt * tela[i].ax
            tela[i].vy += dt * tela[i].ay
            tela[i].vz += dt * tela[i].az

            tela[i].x += dt * tela[i].vx
            tela[i].y += dt * tela[i].vy
            tela[i].z += dt * tela[i].vz
            """
            if i == 0:
                print(i, tela[i].ax, tela[i].vx, dt * tela[i].ax, tela[i].x, dt * tela[i].vx)
            if i == 50:
                print(i, tela[i].ax, tela[i].vx, dt * tela[i].ax, tela[i].x, dt * tela[i].vx)
            if i == 100:
                print(i, tela[i].ax, tela[i].vx, dt * tela[i].ax, tela[i].x, dt * tela[i].vx)
            if i == 150:
                print(i, tela[i].ax, tela[i].vx, dt * tela[i].ax, tela[i].x, dt * tela[i].vx)
            if i == 227:
                print(i, tela[i].ax, tela[i].vx, dt * tela[i].ax, tela[i].x, dt * tela[i].vx)
            """        

        for i in range(N):
            tela[i].ax = 0
            tela[i].ay = 0
            tela[i].az = 0


        print(100.0*t/tmax)


        if t % (1e8 * 365 * 86400) == 0:        
            print(t )
            x1_h = []
            y1_h = []
            z1_h = []

            x1_d = []
            y1_d = []
            z1_d = []
            
            x1_b = []
            y1_b = []
            z1_b = []


            for i in range(N):
                if (tela[i].marker == 'halo'):
                    x1_h.append(tela[i].x)
                    y1_h.append(tela[i].y)
                    z1_h.append(tela[i].z)
                if (tela[i].marker == 'disk'):
                    x1_d.append(tela[i].x)
                    y1_d.append(tela[i].y)
                    z1_d.append(tela[i].z)
                if (tela[i].marker == 'bulge'):
                    x1_b.append(tela[i].x)
                    y1_b.append(tela[i].y)
                    z1_b.append(tela[i].z)

            plt.scatter(x1_h,y1_h)
            plt.scatter(x1_d,y1_d)
            plt.scatter(x1_b,y1_b)
            plt.xlim([-165, 350])
            plt.ylim([-165, 165])
            plt.show()


        t += dt
        #print(i)
    
    plt.scatter(x1_h,y1_h)
    plt.scatter(x1_d,y1_d)
    plt.scatter(x1_b,y1_b)
    plt.show()

    
main()
```
