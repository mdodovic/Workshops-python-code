```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
path = "galaxy.txt"
m,x,y,z,vx,vy,vz = np.loadtxt(path, dtype='float', unpack = True)

m_d = []
m_b = []
m_h = []
x_d = []
x_b = []
x_h = []
y_d = []
y_b = []
y_h = []
z_d = []
z_b = []
z_h = []
vx_d = []
vx_b = []
vx_h = []
vy_d = []
vy_b = []
vy_h = []
vz_d = []
vz_b = []
vz_h = []

for i in range(100):
    m_d.append(m[i])
    x_d.append(x[i])
    y_d.append(y[i])
    z_d.append(z[i])
    vx_d.append(vx[i])
    vy_d.append(vy[i])
    vz_d.append(vz[i])

for i in range (100):
    m_b.append(m[i+100])
    x_b.append(x[i+100])
    y_b.append(y[i+100])
    z_b.append(z[i+100])
    vx_b.append(vx[i+100])
    vy_b.append(vy[i+100])
    vz_b.append(vz[i+100])

for i in range (200):
    m_h.append(m[i+200])
    x_h.append(x[i+200])
    y_h.append(y[i+200])
    z_h.append(z[i+200])
    vx_h.append(vx[i+200])
    vy_h.append(vy[i+200])
    vz_h.append(vz[i+200])
```

```python
class telo():
    def __init__(self, m, x, y, z, vx, vy, vz, ax, ay, az, marker):
        self.m = m
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.ax = ax
        self.ay = ay
        self.az = az
        self.marker = marker


def jedinice_masa(m):
    return m/(2e30)

def jedinice_brzina(v):
    return v*3.24e-17
  
def rastojanje(t1,t2):
    return math.sqrt((t1.x-t2.x)**2+(t1.y-t2.y)**2 + (t1.z-t2.z)**2)

def main():
    
    N = 400
    dt = 5e6 
    G = 4.53e-39
    tmax = 5e9 
    t = 0

    tela = []
    
    for i in range(200):
        tela.append(telo(jedinice_masa(m_h[i]), x_h[i], y_h[i], z_h[i],jedinice_brzina(vx_h[i]), jedinice_brzina(vy_h[i]), jedinice_brzina(vz_h[i]), 0, 0, 0, "halo"))
    for i in range(100):
        tela.append(telo(jedinice_masa(m_d[i]), x_d[i], y_d[i], z_d[i], jedinice_brzina(vx_d[i]), jedinice_brzina(vy_d[i]), jedinice_brzina(vz_d[i]), 0, 0, 0, "disk")) 
    for i in range(100):
        tela.append(telo(jedinice_masa(m_b[i]), x_b[i], y_b[i], z_b[i], jedinice_brzina(vx_b[i]), jedinice_brzina(vy_b[i]), jedinice_brzina(vz_b[i]), 0, 0, 0, "bulge"))

    x1_h = []
    y1_h = []
    z1_h = []

    x1_d = []
    y1_d = []
    z1_d = []
    
    x1_b = []
    y1_b = []
    z1_b = []
    
    while(t<tmax):

        for i in range(N):
            for j in range(N):
                #print(i,j)
                if (i!=j):

                    d = rastojanje(tela[i],tela[j])
                    #print("\nx ",tela[i].x,tela[j].x) 
                    #print("\ny ",tela[i].y,tela[j].y) 
                    #print("\nz ",tela[i].z,tela[j].z) 
                    #print(i,j,d)
                    
                    tela[i].ax += -G * tela[j].m * (tela[i].x-tela[j].x) / ((d)**3)
                    tela[i].ay += -G * tela[j].m * (tela[i].y-tela[j].y) / ((d)**3)
                    tela[i].az += -G * tela[j].m * (tela[i].z-tela[j].z) / ((d)**3)


        for i in range(N):
            tela[i].vx += dt * tela[i].ax
            tela[i].vy += dt * tela[i].ay
            tela[i].vz += dt * tela[i].az

            tela[i].x += dt * tela[i].vx
            tela[i].y += dt * tela[i].vy
            tela[i].z += dt * tela[i].vz

            if (tela[i].marker == 'halo'):
                x1_h.append(tela[i].x)
                y1_h.append(tela[i].y)
                z1_h.append(tela[i].z)
            if (tela[i].marker == 'disk'):
                x1_d.append(tela[i].x)
                y1_d.append(tela[i].y)
                z1_d.append(tela[i].z)
            if (tela[i].marker == 'bulge'):
                x1_b.append(tela[i].x)
                y1_b.append(tela[i].y)
                z1_b.append(tela[i].z)
        

        for i in range(N):
            tela[i].ax = 0
            tela[i].ay = 0
            tela[i].az = 0
        
        print(100.0*t/tmax)
        t += dt
        #print(i)
    
    plt.scatter(x1_h,y1_h)
    plt.scatter(x1_d,y1_d)
    plt.scatter(x1_b,y1_b)
    plt.show()

    
main()
```
